### Purpose is a categorical variable, therefore creating multiple features from the purpose feature.

```python
cat_feats = ['purpose']
```

```python
# Getting the Dummies and adding it into the main dataset.
final_data = pd.get_dummies(loans, columns= cat_feats, drop_first= True)
```

```python
final_data.head()
```

### Train Test Split

```python
from sklearn.cross_validation import train_test_split
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Training Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating Instance
dtree = DecisionTreeClassifier()

# Fitting the model
dtree.fit(X_train, y_train)
```

#### Model Predictions and Evaluations

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

### Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
rfc.fit(X_train, y_train)
```

#### Model Predictions and Evaluations

```python
preds = rfc.predict(X_test)
```

```python
print(classification_report(y_test, preds))
print(confusion_matrix(y_test, preds))
```

#### Overall the Random forest model did better. Both Single decision tree and Random forest did good job for specific class.
